```python

import os
import re
import json
import math
import argparse
from dataclasses import dataclass, asdict, field
from typing import List, Dict, Any, Optional

try:
    from llama_cpp import Llama
except Exception as e:
    raise RuntimeError(
        "llama_cpp is required. Install with: pip install llama-cpp-python"
    ) from e

try:
    from summa import summarizer as summa_summarizer
    from summa import keywords as summa_keywords
except Exception:
    summa_summarizer = None
    summa_keywords = None

try:
    import nltk
    from nltk.tokenize import sent_tokenize as nltk_sent_tokenize, wordpunct_tokenize
except Exception:
    nltk = None
    nltk_sent_tokenize = None
    wordpunct_tokenize = None


# =========================
# Config
# =========================
DEFAULT_GGUF_PATH = os.environ.get("GGUF_PATH", "./model.gguf")
DEFAULT_CTX_SIZE = 4096
DEFAULT_THREADS = max(1, os.cpu_count() or 2)


# =========================
# Data classes
# =========================
@dataclass
class SummaryConfig:
    mode: str = "auto"  # auto | resume | blog | novel | generic
    max_input_chars: int = 120000
    chunk_chars: int = 5000
    chunk_overlap: int = 500
    bullet_count: int = 8
    keyword_count: int = 15
    paragraph_sentences: int = 5
    top_quotes: int = 3
    temperature: float = 0.35
    top_p: float = 0.9
    max_tokens_chunk_summary: int = 280
    max_tokens_final_summary: int = 700


@dataclass
class SummaryOutput:
    detected_mode: str
    title: str
    short_summary: str
    paragraph_summary: str
    hyper_condensed_bullets: List[str] = field(default_factory=list)
    detailed_bullets: List[str] = field(default_factory=list)
    keywords: List[str] = field(default_factory=list)
    named_entities: List[str] = field(default_factory=list)
    resume_highlights: List[str] = field(default_factory=list)
    quotes_or_lines: List[str] = field(default_factory=list)
    chunk_summaries: List[str] = field(default_factory=list)


# =========================
# Model loader
# =========================
class LocalSummarizer:
    def __init__(
        self,
        model_path: str = DEFAULT_GGUF_PATH,
        n_ctx: int = DEFAULT_CTX_SIZE,
        n_threads: int = DEFAULT_THREADS,
    ):
        if not os.path.exists(model_path):
            raise FileNotFoundError(
                f"Model file not found: {model_path}\n"
                f"Set GGUF_PATH or pass --model path/to/model.gguf"
            )

        self.n_ctx = int(n_ctx)
        self.llm = Llama(
            model_path=model_path,
            n_ctx=self.n_ctx,
            n_threads=n_threads,
            verbose=False,
        )

    # -------------------------
    # Core LLM call
    # -------------------------
    def llm_call(
        self,
        prompt: str,
        max_tokens: int = 400,
        temperature: float = 0.35,
        top_p: float = 0.9,
        repeat_penalty: float = 1.08,
        stop: Optional[List[str]] = None,
    ) -> str:
        prompt = self.trim_prompt_to_ctx(prompt, max_tokens=max_tokens)
        out = self.llm(
            prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            repeat_penalty=repeat_penalty,
            stop=stop or [],
        )
        return out["choices"][0]["text"].strip()

    def approximate_token_count(self, text: str) -> int:
        clean = (text or "").strip()
        if not clean:
            return 0
        try:
            return len(self.llm.tokenize(clean.encode("utf-8")))
        except Exception:
            return max(1, int(len(clean) / 3.5))

    def trim_prompt_to_ctx(self, prompt: str, max_tokens: int, reserve_tokens: int = 96) -> str:
        clean = (prompt or "").strip()
        if not clean:
            return ""
        budget = max(256, self.n_ctx - int(max_tokens) - reserve_tokens)
        if self.approximate_token_count(clean) <= budget:
            return clean

        # Head/tail compaction
        while self.approximate_token_count(clean) > budget and len(clean) > 2000:
            keep = int(len(clean) * 0.82)
            head = int(keep * 0.35)
            tail = int(keep * 0.65)
            clean = (
                clean[:head].rstrip()
                + "\n\n[... trimmed for context budget ...]\n\n"
                + clean[-tail:].lstrip()
            )
        return clean

    # =========================
    # Text utilities
    # =========================
    @staticmethod
    def clean_text(text: str) -> str:
        if not text:
            return ""
        text = text.replace("\r\n", "\n").replace("\r", "\n")
        text = re.sub(r"[ \t]+", " ", text)
        text = re.sub(r"\n{3,}", "\n\n", text)
        return text.strip()

    @staticmethod
    def short_text_summary(text: str, max_sentences: int = 3, max_chars: int = 500) -> str:
        parts = [p.strip() for p in re.split(r"(?<=[.!?])\s+", (text or "").strip()) if p.strip()]
        out = " ".join(parts[:max_sentences]) if parts else (text or "").strip()
        return out[:max_chars]

    def safe_sentence_tokenize(self, text: str) -> List[str]:
        clean = (text or "").strip()
        if not clean:
            return []
        if nltk_sent_tokenize is not None:
            try:
                return [s.strip() for s in nltk_sent_tokenize(clean) if s.strip()]
            except Exception:
                pass
        return [s.strip() for s in re.split(r"(?<=[.!?])\s+", clean) if s.strip()]

    @staticmethod
    def safe_word_tokenize(text: str) -> List[str]:
        clean = (text or "").strip()
        if not clean:
            return []
        if wordpunct_tokenize is not None:
            try:
                return [tok for tok in wordpunct_tokenize(clean) if tok and not tok.isspace()]
            except Exception:
                pass
        return re.findall(r"[A-Za-z0-9_\-/']+", clean)

    @staticmethod
    def dedupe_preserve_order(items: List[str]) -> List[str]:
        out = []
        seen = set()
        for item in items:
            key = item.strip().lower()
            if item and key not in seen:
                out.append(item.strip())
                seen.add(key)
        return out

    @staticmethod
    def extract_json_object(text: str) -> Dict[str, Any]:
        if not text:
            return {}
        decoder = json.JSONDecoder()
        candidates = []

        match = re.search(r"\{.*\}", text, re.S)
        if match:
            candidates.append(match.group(0))

        candidates.append(text.strip())

        for candidate in candidates:
            try:
                obj, _ = decoder.raw_decode(candidate)
                if isinstance(obj, dict):
                    return obj
            except Exception:
                continue
        return {}

    # =========================
    # Heuristics
    # =========================
    def detect_mode(self, text: str, forced_mode: str = "auto") -> str:
        if forced_mode and forced_mode != "auto":
            return forced_mode

        low = text.lower()

        resume_score = 0
        blog_score = 0
        novel_score = 0

        resume_markers = [
            "experience", "education", "skills", "projects", "certifications",
            "linkedin", "github", "work history", "professional summary",
            "employment", "intern", "bachelor", "master", "resume"
        ]
        blog_markers = [
            "introduction", "conclusion", "blog", "article", "how to",
            "why", "what makes", "in this post", "let's explore"
        ]
        novel_markers = [
            "chapter", "he said", "she said", "they walked", "the room",
            "suddenly", "whispered", "looked at", "felt", "story"
        ]

        for marker in resume_markers:
            if marker in low:
                resume_score += 2
        for marker in blog_markers:
            if marker in low:
                blog_score += 1
        for marker in novel_markers:
            if marker in low:
                novel_score += 1

        if re.search(r"\b(20\d{2}|19\d{2})\b", text) and re.search(r"\b(engineer|manager|developer|analyst|intern)\b", low):
            resume_score += 2

        if len(re.findall(r"•|- |\* ", text)) > 5:
            resume_score += 2

        if "chapter" in low:
            novel_score += 2

        scores = {
            "resume": resume_score,
            "blog": blog_score,
            "novel": novel_score,
            "generic": 0,
        }
        best = max(scores.items(), key=lambda x: x[1])[0]
        return best if scores[best] > 0 else "generic"

    def extract_title(self, text: str, mode: str) -> str:
        lines = [ln.strip() for ln in text.splitlines() if ln.strip()]
        if not lines:
            return "Untitled Summary"

        if mode == "resume":
            first = lines[0]
            second = lines[1] if len(lines) > 1 else ""
            if len(first.split()) <= 8 and len(second.split()) <= 8:
                return f"{first} — Resume Summary"
            return f"{first[:60]} — Resume Summary"

        if len(lines[0].split()) <= 14:
            return lines[0][:90]

        return f"{mode.title()} Summary"

    def split_into_chunks(self, text: str, chunk_chars: int = 5000, overlap: int = 500) -> List[str]:
        text = self.clean_text(text)
        if len(text) <= chunk_chars:
            return [text]

        paras = text.split("\n\n")
        chunks = []
        current = ""

        for para in paras:
            para = para.strip()
            if not para:
                continue

            if len(current) + len(para) + 2 <= chunk_chars:
                current = current + ("\n\n" if current else "") + para
            else:
                if current:
                    chunks.append(current.strip())
                    tail = current[-overlap:] if overlap > 0 else ""
                    current = (tail + "\n\n" + para).strip()
                else:
                    # large paragraph fallback
                    for i in range(0, len(para), chunk_chars - overlap):
                        piece = para[i:i + chunk_chars]
                        if piece.strip():
                            chunks.append(piece.strip())
                    current = ""

        if current.strip():
            chunks.append(current.strip())

        return chunks

    # =========================
    # Surface extraction
    # =========================
    def nltk_keyword_surface(self, text: str, limit: int = 12) -> List[str]:
        tokens = [
            tok.lower()
            for tok in self.safe_word_tokenize(text)
            if len(tok) > 3 and not tok.isdigit()
        ]
        if not tokens:
            return []

        stop = {
            "this", "that", "with", "from", "have", "your", "their", "about", "there",
            "which", "when", "where", "into", "through", "would", "could", "should",
            "after", "before", "because", "while", "under", "over", "more", "most",
            "also", "than", "then", "very", "such", "been", "were", "they", "them",
            "will", "just", "like", "into", "some", "many", "make", "made", "using",
            "used", "work", "worked"
        }

        counts: Dict[str, int] = {}
        for tok in tokens:
            if tok in stop:
                continue
            counts[tok] = counts.get(tok, 0) + 1

        ranked = sorted(counts.items(), key=lambda x: (-x[1], -len(x[0]), x[0]))
        return [tok for tok, _ in ranked[:limit]]

    def summa_summary_surface(self, text: str, ratio: float = 0.22) -> str:
        clean = (text or "").strip()
        if not clean:
            return ""
        if summa_summarizer is not None and len(clean.split()) > 80:
            try:
                summary = (summa_summarizer.summarize(clean, ratio=ratio) or "").strip()
                if summary:
                    return summary
            except Exception:
                pass
        return self.short_text_summary(clean, max_sentences=4, max_chars=650)

    def summa_keyword_surface(self, text: str, words: int = 12) -> List[str]:
        clean = (text or "").strip()
        if not clean:
            return []
        if summa_keywords is not None:
            try:
                out = summa_keywords.keywords(clean, words=words, split=True)
                if isinstance(out, list):
                    return [str(x).strip().lower() for x in out if str(x).strip()][:words]
            except Exception:
                pass
        return self.nltk_keyword_surface(clean, limit=words)

    def combine_keyword_surfaces(self, *surfaces: List[str], limit: int = 16) -> List[str]:
        out = []
        seen = set()
        for surface in surfaces:
            for item in surface or []:
                token = str(item).strip().lower()
                if token and token not in seen:
                    out.append(token)
                    seen.add(token)
                if len(out) >= limit:
                    return out
        return out

    def extract_named_entities_lightweight(self, text: str, limit: int = 20) -> List[str]:
        candidates = re.findall(
            r"\b[A-Z][a-zA-Z0-9&\-]{1,}(?:\s+[A-Z][a-zA-Z0-9&\-]{1,}){0,3}\b",
            text or ""
        )
        stop = {
            "The", "This", "That", "These", "Those", "And", "But",
            "Chapter", "Summary", "Resume", "Experience", "Education", "Skills"
        }
        cleaned = []
        for item in candidates:
            item = item.strip()
            if item not in stop and len(item) <= 50:
                cleaned.append(item)
        return self.dedupe_preserve_order(cleaned)[:limit]

    def extract_resume_highlights(self, text: str, limit: int = 8) -> List[str]:
        lines = [ln.strip("•-* \t") for ln in text.splitlines() if ln.strip()]
        hits = []
        strong_words = (
            "built", "led", "managed", "improved", "designed", "developed",
            "created", "launched", "optimized", "increased", "reduced",
            "delivered", "implemented", "analyzed", "automated", "engineered"
        )
        for ln in lines:
            low = ln.lower()
            if any(word in low for word in strong_words):
                hits.append(ln)
        return self.dedupe_preserve_order(hits)[:limit]

    def extract_quotes_or_lines(self, text: str, mode: str, limit: int = 3) -> List[str]:
        sentences = self.safe_sentence_tokenize(text)

        if mode == "novel":
            quote_lines = [s for s in sentences if '"' in s or " said " in s.lower() or " whispered " in s.lower()]
            return self.dedupe_preserve_order(quote_lines)[:limit]

        long_lines = [s for s in sentences if 60 <= len(s) <= 180]
        return self.dedupe_preserve_order(long_lines)[:limit]

    # =========================
    # Prompt builders
    # =========================
    def build_chunk_prompt(self, chunk: str, mode: str, config: SummaryConfig) -> str:
        mode_rules = {
            "resume": (
                "Focus on work history, skills, impact, achievements, roles, tools, education, and measurable outcomes."
            ),
            "blog": (
                "Focus on central thesis, main arguments, key examples, practical takeaways, and author intent."
            ),
            "novel": (
                "Focus on plot movement, character actions, emotional tension, themes, setting shifts, and turning points."
            ),
            "generic": (
                "Focus on main ideas, notable details, structure, and important takeaways."
            ),
        }.get(mode, "Focus on the most important information.")

        return f"""
You are a precise summarization engine.

Task:
Summarize the following text chunk into compact JSON.

Mode: {mode}
Instructions:
- Return JSON only.
- Keep wording dense and informative.
- Do not invent facts.
- Keep bullet points concise.
- Extract the core meaning of the chunk.

{mode_rules}

Return JSON with keys:
chunk_summary
bullets
keywords
entities

Rules:
- chunk_summary: 2-4 sentences
- bullets: 3-5 concise bullets
- keywords: 5-10 keywords
- entities: notable people, companies, places, products, schools, books, technologies, or organizations

TEXT CHUNK:
\"\"\"
{chunk}
\"\"\"
""".strip()

    def build_final_prompt(
        self,
        original_text: str,
        mode: str,
        chunk_summaries: List[str],
        chunk_bullets: List[str],
        keywords: List[str],
        entities: List[str],
        config: SummaryConfig,
    ) -> str:
        mode_rules = {
            "resume": """
Target output for resume:
- emphasize role, seniority, strengths, tools, achievements, career arc, and credibility
- infer a strong professional profile only from stated evidence
- keep bullets recruiter-friendly
- include a compact "best fit" style description
""",
            "blog": """
Target output for blog:
- emphasize thesis, arguments, practical lessons, examples, and bottom-line insights
- separate short summary from condensed bullets
""",
            "novel": """
Target output for novel:
- emphasize plot arc, character movement, emotional stakes, themes, and major turning points
- preserve narrative meaning without becoming verbose
""",
            "generic": """
Target output:
- emphasize central ideas, structure, and practical takeaways
- keep summary highly compressed but still readable
""",
        }.get(mode, "")

        joined_chunks = "\n".join(f"- {x}" for x in chunk_summaries[:20])
        joined_bullets = "\n".join(f"- {x}" for x in chunk_bullets[:40])
        joined_keywords = ", ".join(keywords[:25])
        joined_entities = ", ".join(entities[:25])

        excerpt = original_text[:12000]

        return f"""
You are a high-precision summarizer.

Create a final structured summary for the document.

Mode: {mode}

{mode_rules}

Return JSON only with keys:
short_summary
paragraph_summary
hyper_condensed_bullets
detailed_bullets
keywords
named_entities
resume_highlights
quotes_or_lines

Rules:
- short_summary: 1-2 sentences
- paragraph_summary: one tight paragraph
- hyper_condensed_bullets: {config.bullet_count} bullets max, each very compressed
- detailed_bullets: 5-10 bullets
- keywords: keep only the strongest keywords
- named_entities: keep only the strongest entities
- resume_highlights: only meaningful if the text is resume-like, otherwise return []
- quotes_or_lines: include striking lines or representative lines if useful
- no markdown fences
- no explanation outside JSON
- do not invent information not present in the source

Source excerpt:
\"\"\"
{excerpt}
\"\"\"

Chunk summaries:
{joined_chunks}

Chunk bullets:
{joined_bullets}

Keywords:
{joined_keywords}

Entities:
{joined_entities}
""".strip()

    # =========================
    # Main summarize pipeline
    # =========================
    def summarize(self, text: str, config: Optional[SummaryConfig] = None) -> SummaryOutput:
        config = config or SummaryConfig()
        text = self.clean_text(text)[: config.max_input_chars]
        if not text:
            raise ValueError("Input text is empty.")

        detected_mode = self.detect_mode(text, config.mode)
        title = self.extract_title(text, detected_mode)

        chunks = self.split_into_chunks(
            text,
            chunk_chars=config.chunk_chars,
            overlap=config.chunk_overlap,
        )

        chunk_summaries: List[str] = []
        all_bullets: List[str] = []
        all_keywords: List[str] = []
        all_entities: List[str] = []

        # Fast extraction surfaces
        fallback_summary = self.summa_summary_surface(text)
        fallback_keywords = self.combine_keyword_surfaces(
            self.summa_keyword_surface(text, words=config.keyword_count),
            self.nltk_keyword_surface(text, limit=config.keyword_count),
            limit=config.keyword_count,
        )
        fallback_entities = self.extract_named_entities_lightweight(text, limit=20)
        fallback_resume = self.extract_resume_highlights(text, limit=8)
        fallback_quotes = self.extract_quotes_or_lines(text, detected_mode, limit=config.top_quotes)

        # Chunk pass
        for chunk in chunks:
            prompt = self.build_chunk_prompt(chunk, detected_mode, config)
            raw = self.llm_call(
                prompt,
                max_tokens=config.max_tokens_chunk_summary,
                temperature=config.temperature,
                top_p=config.top_p,
                repeat_penalty=1.06,
            )
            js = self.extract_json_object(raw)

            if js:
                chunk_summary = str(js.get("chunk_summary", "")).strip()
                bullets = js.get("bullets", [])
                keywords = js.get("keywords", [])
                entities = js.get("entities", [])

                if chunk_summary:
                    chunk_summaries.append(chunk_summary)

                if isinstance(bullets, list):
                    all_bullets.extend([str(x).strip() for x in bullets if str(x).strip()])

                if isinstance(keywords, list):
                    all_keywords.extend([str(x).strip().lower() for x in keywords if str(x).strip()])

                if isinstance(entities, list):
                    all_entities.extend([str(x).strip() for x in entities if str(x).strip()])
            else:
                # Fallback chunk summary if JSON fails
                chunk_summaries.append(self.summa_summary_surface(chunk))
                all_keywords.extend(self.nltk_keyword_surface(chunk, limit=8))
                all_entities.extend(self.extract_named_entities_lightweight(chunk, limit=8))

        chunk_summaries = self.dedupe_preserve_order(chunk_summaries)
        all_bullets = self.dedupe_preserve_order(all_bullets)
        all_keywords = self.dedupe_preserve_order(all_keywords + fallback_keywords)
        all_entities = self.dedupe_preserve_order(all_entities + fallback_entities)

        # Final synthesis pass
        final_prompt = self.build_final_prompt(
            original_text=text,
            mode=detected_mode,
            chunk_summaries=chunk_summaries,
            chunk_bullets=all_bullets,
            keywords=all_keywords,
            entities=all_entities,
            config=config,
        )

        raw_final = self.llm_call(
            final_prompt,
            max_tokens=config.max_tokens_final_summary,
            temperature=max(0.2, config.temperature),
            top_p=config.top_p,
            repeat_penalty=1.08,
        )
        js_final = self.extract_json_object(raw_final)

        if js_final:
            short_summary = str(js_final.get("short_summary", "")).strip() or self.short_text_summary(fallback_summary, 2, 260)
            paragraph_summary = str(js_final.get("paragraph_summary", "")).strip() or fallback_summary

            hyper_condensed_bullets = js_final.get("hyper_condensed_bullets", [])
            detailed_bullets = js_final.get("detailed_bullets", [])
            keywords = js_final.get("keywords", [])
            named_entities = js_final.get("named_entities", [])
            resume_highlights = js_final.get("resume_highlights", [])
            quotes_or_lines = js_final.get("quotes_or_lines", [])

            if not isinstance(hyper_condensed_bullets, list):
                hyper_condensed_bullets = []
            if not isinstance(detailed_bullets, list):
                detailed_bullets = []
            if not isinstance(keywords, list):
                keywords = []
            if not isinstance(named_entities, list):
                named_entities = []
            if not isinstance(resume_highlights, list):
                resume_highlights = []
            if not isinstance(quotes_or_lines, list):
                quotes_or_lines = []
        else:
            short_summary = self.short_text_summary(fallback_summary, 2, 260)
            paragraph_summary = fallback_summary
            hyper_condensed_bullets = all_bullets[: config.bullet_count]
            detailed_bullets = all_bullets[:10]
            keywords = all_keywords[: config.keyword_count]
            named_entities = all_entities[:15]
            resume_highlights = fallback_resume if detected_mode == "resume" else []
            quotes_or_lines = fallback_quotes

        # Safety fallback filling
        if not hyper_condensed_bullets:
            hyper_condensed_bullets = all_bullets[: config.bullet_count]
        if not detailed_bullets:
            detailed_bullets = all_bullets[:10]
        if not keywords:
            keywords = fallback_keywords[: config.keyword_count]
        if not named_entities:
            named_entities = fallback_entities[:15]
        if detected_mode == "resume" and not resume_highlights:
            resume_highlights = fallback_resume
        if not quotes_or_lines:
            quotes_or_lines = fallback_quotes

        return SummaryOutput(
            detected_mode=detected_mode,
            title=title,
            short_summary=short_summary,
            paragraph_summary=paragraph_summary,
            hyper_condensed_bullets=self.dedupe_preserve_order(
                [str(x).strip() for x in hyper_condensed_bullets if str(x).strip()]
            )[: config.bullet_count],
            detailed_bullets=self.dedupe_preserve_order(
                [str(x).strip() for x in detailed_bullets if str(x).strip()]
            )[:10],
            keywords=self.dedupe_preserve_order(
                [str(x).strip() for x in keywords if str(x).strip()]
            )[: config.keyword_count],
            named_entities=self.dedupe_preserve_order(
                [str(x).strip() for x in named_entities if str(x).strip()]
            )[:15],
            resume_highlights=self.dedupe_preserve_order(
                [str(x).strip() for x in resume_highlights if str(x).strip()]
            )[:8],
            quotes_or_lines=self.dedupe_preserve_order(
                [str(x).strip() for x in quotes_or_lines if str(x).strip()]
            )[: config.top_quotes],
            chunk_summaries=chunk_summaries[:20],
        )

    # =========================
    # Renderers
    # =========================
    @staticmethod
    def render_markdown(result: SummaryOutput) -> str:
        parts = [
            f"# {result.title}",
            "",
            f"**Detected mode:** {result.detected_mode}",
            "",
            "## Short Summary",
            result.short_summary,
            "",
            "## Paragraph Summary",
            result.paragraph_summary,
            "",
            "## Hyper-Condensed Bullets",
        ]
        parts.extend([f"- {x}" for x in result.hyper_condensed_bullets])

        if result.detailed_bullets:
            parts.append("")
            parts.append("## Detailed Bullets")
            parts.extend([f"- {x}" for x in result.detailed_bullets])

        if result.resume_highlights:
            parts.append("")
            parts.append("## Resume Highlights")
            parts.extend([f"- {x}" for x in result.resume_highlights])

        if result.keywords:
            parts.append("")
            parts.append("## Keywords")
            parts.append(", ".join(result.keywords))

        if result.named_entities:
            parts.append("")
            parts.append("## Named Entities")
            parts.append(", ".join(result.named_entities))

        if result.quotes_or_lines:
            parts.append("")
            parts.append("## Representative Lines")
            parts.extend([f"- {x}" for x in result.quotes_or_lines])

        return "\n".join(parts).strip()

    @staticmethod
    def render_console(result: SummaryOutput) -> str:
        lines = [
            f"TITLE: {result.title}",
            f"MODE: {result.detected_mode}",
            "",
            "SHORT SUMMARY:",
            result.short_summary,
            "",
            "PARAGRAPH SUMMARY:",
            result.paragraph_summary,
            "",
            "HYPER-CONDENSED BULLETS:",
        ]
        lines.extend([f"- {x}" for x in result.hyper_condensed_bullets])

        if result.resume_highlights:
            lines.append("")
            lines.append("RESUME HIGHLIGHTS:")
            lines.extend([f"- {x}" for x in result.resume_highlights])

        if result.keywords:
            lines.append("")
            lines.append("KEYWORDS:")
            lines.append(", ".join(result.keywords))

        return "\n".join(lines)


# =========================
# File helpers
# =========================
def load_text_file(path: str) -> str:
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        return f.read()


def save_text_file(path: str, content: str):
    with open(path, "w", encoding="utf-8") as f:
        f.write(content)


def save_json_file(path: str, payload: Dict[str, Any]):
    with open(path, "w", encoding="utf-8") as f:
        json.dump(payload, f, ensure_ascii=False, indent=2)


# =========================
# CLI
# =========================
def build_arg_parser() -> argparse.ArgumentParser:
    parser = argparse.ArgumentParser(
        description="Local resume/blog/novel summarization system using llama.cpp"
    )
    parser.add_argument("--input", type=str, help="Path to input text file")
    parser.add_argument("--text", type=str, default="", help="Raw text input")
    parser.add_argument("--model", type=str, default=DEFAULT_GGUF_PATH, help="Path to GGUF model")
    parser.add_argument("--mode", type=str, default="auto", choices=["auto", "resume", "blog", "novel", "generic"])
    parser.add_argument("--ctx", type=int, default=DEFAULT_CTX_SIZE)
    parser.add_argument("--threads", type=int, default=DEFAULT_THREADS)
    parser.add_argument("--chunk_chars", type=int, default=5000)
    parser.add_argument("--chunk_overlap", type=int, default=500)
    parser.add_argument("--bullet_count", type=int, default=8)
    parser.add_argument("--keyword_count", type=int, default=15)
    parser.add_argument("--paragraph_sentences", type=int, default=5)
    parser.add_argument("--temperature", type=float, default=0.35)
    parser.add_argument("--top_p", type=float, default=0.9)
    parser.add_argument("--out_md", type=str, default="")
    parser.add_argument("--out_json", type=str, default="")
    return parser


def main():
    parser = build_arg_parser()
    args = parser.parse_args()

    if not args.input and not args.text.strip():
        raise ValueError("Provide either --input path/to/file.txt or --text 'your text here'")

    input_text = args.text.strip()
    if args.input:
        input_text = load_text_file(args.input)

    config = SummaryConfig(
        mode=args.mode,
        chunk_chars=args.chunk_chars,
        chunk_overlap=args.chunk_overlap,
        bullet_count=args.bullet_count,
        keyword_count=args.keyword_count,
        paragraph_sentences=args.paragraph_sentences,
        temperature=args.temperature,
        top_p=args.top_p,
    )

    engine = LocalSummarizer(
        model_path=args.model,
        n_ctx=args.ctx,
        n_threads=args.threads,
    )

    result = engine.summarize(input_text, config=config)

    console_view = engine.render_console(result)
    print(console_view)

    if args.out_md:
        save_text_file(args.out_md, engine.render_markdown(result))

    if args.out_json:
        save_json_file(args.out_json, asdict(result))


if __name__ == "__main__":
    main()
```
