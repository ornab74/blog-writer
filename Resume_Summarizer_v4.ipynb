```python
# =========================
# CELL 5 — Summarizer system
# =========================

import os
import re
import json
from dataclasses import dataclass, asdict, field
from typing import List, Dict, Any, Optional

from llama_cpp import Llama

try:
    from summa import summarizer as summa_summarizer
    from summa import keywords as summa_keywords
except Exception:
    summa_summarizer = None
    summa_keywords = None

try:
    from nltk.tokenize import sent_tokenize as nltk_sent_tokenize, wordpunct_tokenize
except Exception:
    nltk_sent_tokenize = None
    wordpunct_tokenize = None


@dataclass
class SummaryConfig:
    mode: str = "auto"   # auto | resume | blog | novel | generic
    max_input_chars: int = 120000
    chunk_chars: int = 4500
    chunk_overlap: int = 400
    bullet_count: int = 8
    keyword_count: int = 15
    paragraph_sentences: int = 5
    top_quotes: int = 3
    temperature: float = 0.30
    top_p: float = 0.90
    max_tokens_chunk_summary: int = 260
    max_tokens_final_summary: int = 700


@dataclass
class SummaryOutput:
    detected_mode: str
    title: str
    short_summary: str
    paragraph_summary: str
    hyper_condensed_bullets: List[str] = field(default_factory=list)
    detailed_bullets: List[str] = field(default_factory=list)
    keywords: List[str] = field(default_factory=list)
    named_entities: List[str] = field(default_factory=list)
    resume_highlights: List[str] = field(default_factory=list)
    quotes_or_lines: List[str] = field(default_factory=list)
    chunk_summaries: List[str] = field(default_factory=list)


class LocalSummarizer:
    def __init__(self, model_path: str, n_ctx: int = 4096, n_threads: int = 2):
        if not os.path.exists(model_path):
            raise FileNotFoundError(f"Model file not found: {model_path}")
        self.n_ctx = int(n_ctx)
        self.llm = Llama(
            model_path=model_path,
            n_ctx=n_ctx,
            n_threads=n_threads,
            verbose=False,
        )

    def llm_call(
        self,
        prompt: str,
        max_tokens: int = 400,
        temperature: float = 0.3,
        top_p: float = 0.9,
        repeat_penalty: float = 1.08,
        stop: Optional[List[str]] = None,
    ) -> str:
        prompt = self.trim_prompt_to_ctx(prompt, max_tokens=max_tokens)
        out = self.llm(
            prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            repeat_penalty=repeat_penalty,
            stop=stop or [],
        )
        return out["choices"][0]["text"].strip()

    def approximate_token_count(self, text: str) -> int:
        clean = (text or "").strip()
        if not clean:
            return 0
        try:
            return len(self.llm.tokenize(clean.encode("utf-8")))
        except Exception:
            return max(1, int(len(clean) / 3.5))

    def trim_prompt_to_ctx(self, prompt: str, max_tokens: int, reserve_tokens: int = 96) -> str:
        clean = (prompt or "").strip()
        if not clean:
            return ""
        budget = max(256, self.n_ctx - int(max_tokens) - reserve_tokens)
        if self.approximate_token_count(clean) <= budget:
            return clean

        while self.approximate_token_count(clean) > budget and len(clean) > 2000:
            keep = int(len(clean) * 0.82)
            head = int(keep * 0.35)
            tail = int(keep * 0.65)
            clean = clean[:head].rstrip() + "\n\n[...trimmed...]\n\n" + clean[-tail:].lstrip()
        return clean

    @staticmethod
    def clean_text(text: str) -> str:
        if not text:
            return ""
        text = text.replace("\r\n", "\n").replace("\r", "\n")
        text = re.sub(r"[ \t]+", " ", text)
        text = re.sub(r"\n{3,}", "\n\n", text)
        return text.strip()

    def safe_sentence_tokenize(self, text: str) -> List[str]:
        clean = (text or "").strip()
        if not clean:
            return []
        if nltk_sent_tokenize is not None:
            try:
                return [s.strip() for s in nltk_sent_tokenize(clean) if s.strip()]
            except Exception:
                pass
        return [s.strip() for s in re.split(r"(?<=[.!?])\s+", clean) if s.strip()]

    @staticmethod
    def safe_word_tokenize(text: str) -> List[str]:
        clean = (text or "").strip()
        if not clean:
            return []
        if wordpunct_tokenize is not None:
            try:
                return [tok for tok in wordpunct_tokenize(clean) if tok and not tok.isspace()]
            except Exception:
                pass
        return re.findall(r"[A-Za-z0-9_\-/']+", clean)

    @staticmethod
    def dedupe_preserve_order(items: List[str]) -> List[str]:
        out = []
        seen = set()
        for item in items:
            key = item.strip().lower()
            if item and key not in seen:
                out.append(item.strip())
                seen.add(key)
        return out

    @staticmethod
    def extract_json_object(text: str) -> Dict[str, Any]:
        if not text:
            return {}
        decoder = json.JSONDecoder()
        candidates = []
        match = re.search(r"\{.*\}", text, re.S)
        if match:
            candidates.append(match.group(0))
        candidates.append(text.strip())

        for candidate in candidates:
            try:
                obj, _ = decoder.raw_decode(candidate)
                if isinstance(obj, dict):
                    return obj
            except Exception:
                continue
        return {}

    def detect_mode(self, text: str, forced_mode: str = "auto") -> str:
        if forced_mode and forced_mode != "auto":
            return forced_mode

        low = text.lower()
        resume_score = 0
        blog_score = 0
        novel_score = 0

        for marker in ["experience", "education", "skills", "projects", "certifications", "resume", "professional summary"]:
            if marker in low:
                resume_score += 2

        for marker in ["blog", "article", "introduction", "conclusion", "how to", "in this post"]:
            if marker in low:
                blog_score += 1

        for marker in ["chapter", "he said", "she said", "whispered", "looked at", "story"]:
            if marker in low:
                novel_score += 1

        if len(re.findall(r"•|- |\* ", text)) > 5:
            resume_score += 2

        scores = {
            "resume": resume_score,
            "blog": blog_score,
            "novel": novel_score,
            "generic": 0,
        }
        best = max(scores.items(), key=lambda x: x[1])[0]
        return best if scores[best] > 0 else "generic"

    def extract_title(self, text: str, mode: str) -> str:
        lines = [ln.strip() for ln in text.splitlines() if ln.strip()]
        if not lines:
            return "Untitled Summary"
        if mode == "resume":
            return f"{lines[0]} — Resume Summary"
        if len(lines[0].split()) <= 14:
            return lines[0][:90]
        return f"{mode.title()} Summary"

    def split_into_chunks(self, text: str, chunk_chars: int = 4500, overlap: int = 400) -> List[str]:
        text = self.clean_text(text)
        if len(text) <= chunk_chars:
            return [text]

        paras = text.split("\n\n")
        chunks = []
        current = ""

        for para in paras:
            para = para.strip()
            if not para:
                continue
            if len(current) + len(para) + 2 <= chunk_chars:
                current = current + ("\n\n" if current else "") + para
            else:
                if current:
                    chunks.append(current.strip())
                    tail = current[-overlap:] if overlap > 0 else ""
                    current = (tail + "\n\n" + para).strip()
                else:
                    chunks.append(para[:chunk_chars])

        if current.strip():
            chunks.append(current.strip())

        return chunks

    def short_text_summary(self, text: str, max_sentences: int = 3, max_chars: int = 500) -> str:
        parts = [p.strip() for p in re.split(r"(?<=[.!?])\s+", (text or "").strip()) if p.strip()]
        out = " ".join(parts[:max_sentences]) if parts else (text or "").strip()
        return out[:max_chars]

    def nltk_keyword_surface(self, text: str, limit: int = 12) -> List[str]:
        tokens = [tok.lower() for tok in self.safe_word_tokenize(text) if len(tok) > 3 and not tok.isdigit()]
        stop = {
            "this", "that", "with", "from", "have", "your", "their", "about", "there",
            "which", "when", "where", "into", "through", "would", "could", "should",
            "after", "before", "because", "while", "under", "over", "more", "most",
            "also", "than", "then", "very", "such", "been", "were", "they", "them",
            "will", "just", "like", "some", "many", "make", "made", "using", "used"
        }
        counts = {}
        for tok in tokens:
            if tok in stop:
                continue
            counts[tok] = counts.get(tok, 0) + 1
        ranked = sorted(counts.items(), key=lambda x: (-x[1], -len(x[0]), x[0]))
        return [tok for tok, _ in ranked[:limit]]

    def summa_summary_surface(self, text: str, ratio: float = 0.22) -> str:
        clean = (text or "").strip()
        if not clean:
            return ""
        if summa_summarizer is not None and len(clean.split()) > 80:
            try:
                summary = (summa_summarizer.summarize(clean, ratio=ratio) or "").strip()
                if summary:
                    return summary
            except Exception:
                pass
        return self.short_text_summary(clean, max_sentences=4, max_chars=650)

    def summa_keyword_surface(self, text: str, words: int = 12) -> List[str]:
        clean = (text or "").strip()
        if not clean:
            return []
        if summa_keywords is not None:
            try:
                out = summa_keywords.keywords(clean, words=words, split=True)
                if isinstance(out, list):
                    return [str(x).strip().lower() for x in out if str(x).strip()][:words]
            except Exception:
                pass
        return self.nltk_keyword_surface(clean, limit=words)

    def combine_keyword_surfaces(self, *surfaces: List[str], limit: int = 16) -> List[str]:
        out = []
        seen = set()
        for surface in surfaces:
            for item in surface or []:
                token = str(item).strip().lower()
                if token and token not in seen:
                    out.append(token)
                    seen.add(token)
                if len(out) >= limit:
                    return out
        return out

    def extract_named_entities_lightweight(self, text: str, limit: int = 20) -> List[str]:
        candidates = re.findall(r"\b[A-Z][a-zA-Z0-9&\-]{1,}(?:\s+[A-Z][a-zA-Z0-9&\-]{1,}){0,3}\b", text or "")
        stop = {"The", "This", "That", "These", "Those", "And", "But", "Experience", "Education", "Skills"}
        cleaned = []
        for item in candidates:
            item = item.strip()
            if item not in stop and len(item) <= 50:
                cleaned.append(item)
        return self.dedupe_preserve_order(cleaned)[:limit]

    def extract_resume_highlights(self, text: str, limit: int = 8) -> List[str]:
        lines = [ln.strip("•-* \t") for ln in text.splitlines() if ln.strip()]
        strong_words = (
            "built", "led", "managed", "improved", "designed", "developed",
            "created", "launched", "optimized", "increased", "reduced",
            "delivered", "implemented", "analyzed", "automated", "engineered"
        )
        hits = []
        for ln in lines:
            low = ln.lower()
            if any(word in low for word in strong_words):
                hits.append(ln)
        return self.dedupe_preserve_order(hits)[:limit]

    def extract_quotes_or_lines(self, text: str, mode: str, limit: int = 3) -> List[str]:
        sentences = self.safe_sentence_tokenize(text)
        if mode == "novel":
            quote_lines = [s for s in sentences if '"' in s or " said " in s.lower()]
            return self.dedupe_preserve_order(quote_lines)[:limit]
        long_lines = [s for s in sentences if 60 <= len(s) <= 180]
        return self.dedupe_preserve_order(long_lines)[:limit]

    def build_chunk_prompt(self, chunk: str, mode: str) -> str:
        return f"""
You are a precise summarization engine.

Summarize the following text chunk into JSON only.

Mode: {mode}

Return JSON with keys:
chunk_summary
bullets
keywords
entities

Rules:
- chunk_summary: 2 to 4 sentences
- bullets: 3 to 5 concise bullets
- keywords: 5 to 10 keywords
- entities: notable people, companies, places, tools, schools, or organizations
- do not invent facts
- output JSON only

TEXT:
\"\"\"
{chunk}
\"\"\"
""".strip()

    def build_final_prompt(
        self,
        original_text: str,
        mode: str,
        chunk_summaries: List[str],
        chunk_bullets: List[str],
        keywords: List[str],
        entities: List[str],
        config: SummaryConfig,
    ) -> str:
        return f"""
You are a high-precision summarizer.

Create a final structured summary for the document.

Mode: {mode}

Return JSON only with keys:
short_summary
paragraph_summary
hyper_condensed_bullets
detailed_bullets
keywords
named_entities
resume_highlights
quotes_or_lines

Rules:
- short_summary: 1 to 2 sentences
- paragraph_summary: one tight paragraph
- hyper_condensed_bullets: max {config.bullet_count} bullets
- detailed_bullets: 5 to 10 bullets
- keywords: strongest keywords only
- named_entities: strongest entities only
- resume_highlights: only meaningful for resumes, else []
- quotes_or_lines: representative lines if useful
- do not invent information
- output JSON only

Source excerpt:
\"\"\"
{original_text[:12000]}
\"\"\"

Chunk summaries:
{chr(10).join('- ' + x for x in chunk_summaries[:20])}

Chunk bullets:
{chr(10).join('- ' + x for x in chunk_bullets[:40])}

Keywords:
{", ".join(keywords[:25])}

Entities:
{", ".join(entities[:25])}
""".strip()

    def summarize(self, text: str, config: Optional[SummaryConfig] = None) -> SummaryOutput:
        config = config or SummaryConfig()
        text = self.clean_text(text)[: config.max_input_chars]
        if not text:
            raise ValueError("Input text is empty.")

        detected_mode = self.detect_mode(text, config.mode)
        title = self.extract_title(text, detected_mode)
        chunks = self.split_into_chunks(text, config.chunk_chars, config.chunk_overlap)

        chunk_summaries = []
        all_bullets = []
        all_keywords = []
        all_entities = []

        fallback_summary = self.summa_summary_surface(text)
        fallback_keywords = self.combine_keyword_surfaces(
            self.summa_keyword_surface(text, words=config.keyword_count),
            self.nltk_keyword_surface(text, limit=config.keyword_count),
            limit=config.keyword_count,
        )
        fallback_entities = self.extract_named_entities_lightweight(text, limit=20)
        fallback_resume = self.extract_resume_highlights(text, limit=8)
        fallback_quotes = self.extract_quotes_or_lines(text, detected_mode, limit=config.top_quotes)

        for chunk in chunks:
            raw = self.llm_call(
                self.build_chunk_prompt(chunk, detected_mode),
                max_tokens=config.max_tokens_chunk_summary,
                temperature=config.temperature,
                top_p=config.top_p,
                repeat_penalty=1.06,
            )
            js = self.extract_json_object(raw)
            if js:
                chunk_summary = str(js.get("chunk_summary", "")).strip()
                bullets = js.get("bullets", [])
                keywords = js.get("keywords", [])
                entities = js.get("entities", [])

                if chunk_summary:
                    chunk_summaries.append(chunk_summary)
                if isinstance(bullets, list):
                    all_bullets.extend([str(x).strip() for x in bullets if str(x).strip()])
                if isinstance(keywords, list):
                    all_keywords.extend([str(x).strip().lower() for x in keywords if str(x).strip()])
                if isinstance(entities, list):
                    all_entities.extend([str(x).strip() for x in entities if str(x).strip()])
            else:
                chunk_summaries.append(self.summa_summary_surface(chunk))
                all_keywords.extend(self.nltk_keyword_surface(chunk, limit=8))
                all_entities.extend(self.extract_named_entities_lightweight(chunk, limit=8))

        chunk_summaries = self.dedupe_preserve_order(chunk_summaries)
        all_bullets = self.dedupe_preserve_order(all_bullets)
        all_keywords = self.dedupe_preserve_order(all_keywords + fallback_keywords)
        all_entities = self.dedupe_preserve_order(all_entities + fallback_entities)

        raw_final = self.llm_call(
            self.build_final_prompt(
                original_text=text,
                mode=detected_mode,
                chunk_summaries=chunk_summaries,
                chunk_bullets=all_bullets,
                keywords=all_keywords,
                entities=all_entities,
                config=config,
            ),
            max_tokens=config.max_tokens_final_summary,
            temperature=max(0.2, config.temperature),
            top_p=config.top_p,
            repeat_penalty=1.08,
        )

        js_final = self.extract_json_object(raw_final)

        if js_final:
            short_summary = str(js_final.get("short_summary", "")).strip() or self.short_text_summary(fallback_summary, 2, 260)
            paragraph_summary = str(js_final.get("paragraph_summary", "")).strip() or fallback_summary
            hyper_condensed_bullets = js_final.get("hyper_condensed_bullets", [])
            detailed_bullets = js_final.get("detailed_bullets", [])
            keywords = js_final.get("keywords", [])
            named_entities = js_final.get("named_entities", [])
            resume_highlights = js_final.get("resume_highlights", [])
            quotes_or_lines = js_final.get("quotes_or_lines", [])
        else:
            short_summary = self.short_text_summary(fallback_summary, 2, 260)
            paragraph_summary = fallback_summary
            hyper_condensed_bullets = all_bullets[: config.bullet_count]
            detailed_bullets = all_bullets[:10]
            keywords = all_keywords[: config.keyword_count]
            named_entities = all_entities[:15]
            resume_highlights = fallback_resume if detected_mode == "resume" else []
            quotes_or_lines = fallback_quotes

        return SummaryOutput(
            detected_mode=detected_mode,
            title=title,
            short_summary=short_summary,
            paragraph_summary=paragraph_summary,
            hyper_condensed_bullets=self.dedupe_preserve_order([str(x).strip() for x in hyper_condensed_bullets if str(x).strip()])[: config.bullet_count],
            detailed_bullets=self.dedupe_preserve_order([str(x).strip() for x in detailed_bullets if str(x).strip()])[:10],
            keywords=self.dedupe_preserve_order([str(x).strip() for x in keywords if str(x).strip()])[: config.keyword_count],
            named_entities=self.dedupe_preserve_order([str(x).strip() for x in named_entities if str(x).strip()])[:15],
            resume_highlights=self.dedupe_preserve_order([str(x).strip() for x in resume_highlights if str(x).strip()])[:8],
            quotes_or_lines=self.dedupe_preserve_order([str(x).strip() for x in quotes_or_lines if str(x).strip()])[: config.top_quotes],
            chunk_summaries=chunk_summaries[:20],
        )

    @staticmethod
    def render_markdown(result: SummaryOutput) -> str:
        parts = [
            f"# {result.title}",
            "",
            f"**Detected mode:** {result.detected_mode}",
            "",
            "## Short Summary",
            result.short_summary,
            "",
            "## Paragraph Summary",
            result.paragraph_summary,
            "",
            "## Hyper-Condensed Bullets",
        ]
        parts.extend([f"- {x}" for x in result.hyper_condensed_bullets])

        if result.detailed_bullets:
            parts.append("")
            parts.append("## Detailed Bullets")
            parts.extend([f"- {x}" for x in result.detailed_bullets])

        if result.resume_highlights:
            parts.append("")
            parts.append("## Resume Highlights")
            parts.extend([f"- {x}" for x in result.resume_highlights])

        if result.keywords:
            parts.append("")
            parts.append("## Keywords")
            parts.append(", ".join(result.keywords))

        if result.named_entities:
            parts.append("")
            parts.append("## Named Entities")
            parts.append(", ".join(result.named_entities))

        if result.quotes_or_lines:
            parts.append("")
            parts.append("## Representative Lines")
            parts.extend([f"- {x}" for x in result.quotes_or_lines])

        return "\n".join(parts).strip()
```

```python
# =========================
# CELL 6 — Run summarization
# Change mode to: auto / resume / blog / novel / generic
# =========================

config = SummaryConfig(
    mode="resume",
    bullet_count=8,
    keyword_count=15,
    temperature=0.25,
    top_p=0.9,
)

engine = LocalSummarizer(
    model_path=GGUF_PATH,
    n_ctx=CTX_SIZE,
    n_threads=THREADS,
)

result = engine.summarize(INPUT_TEXT, config=config)

print("TITLE:", result.title)
print("MODE:", result.detected_mode)
print("\nSHORT SUMMARY:\n")
print(result.short_summary)
print("\nPARAGRAPH SUMMARY:\n")
print(result.paragraph_summary)

print("\nHYPER-CONDENSED BULLETS:")
for item in result.hyper_condensed_bullets:
    print("-", item)

if result.resume_highlights:
    print("\nRESUME HIGHLIGHTS:")
    for item in result.resume_highlights:
        print("-", item)

print("\nKEYWORDS:")
print(", ".join(result.keywords))
```
